```python
import pandas as pd
df=pd.read_csv("CC GENERAL.csv", encoding = "iso.8859-1")
```

```python
df.head()
```

```python
df=df.drop('CUST_ID',axis=1)

df
```

```python
df.isnull().sum()
```

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean(), inplace=True)
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
clust_label=model.fit_predict(df)
```

```python
#3. Tracez les clusters.
agglomerative=pd.DataFrame(clust_label)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_subplot(111)
scatter=ax.scatter(df['CASH_ADVANCE'],df['PAYMENTS'],c=agglomerative[0],s=50)
ax.set_title('Aglomerative Clustering')
ax.set_xlabel('CASH_ADVANCE')
ax.set_ylabel('PAYMENTS')
plt.colorbar(scatter)
```

```python
#4 Tracez le dendrogramme. Utilisez le clustering k-means
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10,7))
plt.title('dendrogramme')
dend=shc.dendrogram(shc.linkage(df,method='complete'))
```

```python
#5 Essayez différentes valeurs de k et sélectionnez la meilleure.
from sklearn.cluster import KMeans
Kmeans=KMeans(n_clusters=5,random_state=0)
Kmeans.fit(df)
```

```python
#étiquettes K-means
labels=pd.DataFrame(Kmeans.labels_)
labels
```

```python
#centroides K-means
Kmeans.predict(df)
print(Kmeans.cluster_centers_)
```

```python
#meilleure valeur de k
Sum_of_squared_distance=[]
for k in range(1, 15):
    km=KMeans(n_clusters=k)
    km=km.fit(df)
    Sum_of_squared_distance.append(km.inertia_)
```

```python
Sum_of_squared_distance
```
